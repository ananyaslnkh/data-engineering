```python
df = pd.read_csv('world_population_by_country_2023_modified.csv')
```

```python
print(df)
```

```python
print(df.isnull().sum())
```

```python
drop_all = df.dropna()
print(drop_all.head(15))
```

```python
drop_row = df.dropna(how='all')
print(drop_row.head(15))
```

```python
drop_more_than_2 = df.dropna(thresh=df.shape[1] - 2)
print(drop_more_than_2.head(10))
```

```python
drop_specific = df.dropna(subset=['net_migrants'])
print(drop_specific.head(15))
```

```python
df.fillna(0)
```

```python
mean = df['population_urban'].mean()
print(mean)
df['population_urban'].fillna(mean, inplace=True)
print(df)
```

```python
median = df['median_age'].median()
print(median)
df['median_age'].fillna(median, inplace=True)
print(df)
```

```python
mode = df['population_urban'].mode()
print(mode)
df['population_urban'].fillna(mode, inplace=True)
print(df)
```

```python
duplicates = df.duplicated()
print(duplicates)
```

```python
df.drop_duplicates()
```
