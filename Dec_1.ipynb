```python
# filling missing values using fillna()
# Here, missing values in population_urban column will be filled by 3
dataset['population_urban'].fillna(3,inplace=True)
dataset.head(20)
```

```python
# filling missing value using median()
median = dataset['world_share'].median()
print(median)
dataset["world_share"].fillna(median, inplace=True)
print(dataset.describe())
print("\n")
print(dataset)
```

```python
# fill missing value using mean()
my_mean = dataset["fertility_rate"].mean()
print(my_mean)
dataset["fertility_rate"].fillna(my_mean, inplace=True)
print(dataset.describe())
print("\n")
print(dataset)
```
